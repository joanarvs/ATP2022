### TPC4: Aplicação para Gerir um Cinema

Suponha que está a desenvolver uma aplicacão para gestão de um conjunto de salas de cinema de um centro comercial. 
Nesse centro comercial existem algumas salas de cinema (que poderão estar a exibir filmes ou não), cada sala tem uma determinada 
lotação, uma lista com a referência dos bilhetes vendidos (lugares ocupados; cada lugar é identificado por um número inteiro), e cada sala tem um filme associado.

Considera a seguinte sugestão para o modelo dos cinemas:
```
Cinema = [Sala]
Sala = (nlugares, Vendidos, filme)
nlugares = Int
filme = String 
Vendidos = [Int]
```
  
Que poderá ser usado num programa da seguinte forma:
```
sala1 = (150, [], "Twilight")
sala2 = (200, [], "Hannibal")
cinema1 = []
...
cinema1 = inserirSala(cinema1,sala1)
cinema1 = inserirSala(cinema1,sala2)
...
listar(cinema1)
...

if(disponivel(cinema1, "Twilight", 17 )):
  cinema1 = vendebilhete(cinema1, "Twilight", 17 )
...
listardisponibilidades(cinema1)
...
```

Especifique as funções utilizadas no exemplo:

1. `listar( cinema )` - que lista no monitor todos os filmes que estão em exibição nas salas do cinema passado como argumento;
2. `disponivel( cinema, filme, lugar )` - que dá como resultado **False** se o lugar lugar já estiver ocupado na sala onde o filme está a ser exibido e dará como resultado **True** se o inverso acontecer;
3. `vendebilhete( cinema, filme, lugar )` - que dá como resultado um novo cinema resultante de acrescentar o lugar à lista dos lugares ocupados, na sala onde está a ser exibido o filme;
4. `listardisponibilidades( cinema )` - que, para um dado cinema, lista no monitor para cada sala, o filme que está a ser exibido e o total de lugares disponíveis nessa sala (número de lugares na sala menos o número de lugares ocupados);
5. `inserirSala( cinema, sala )` - que acrescenta uma sala nova a um cinema (devendo verificar se a sala já existe);
6. Acrescente todas as outras funcionalidades que achar necessárias;
7. À semelhança do exercício 3, construa uma aplicação com um menu de interface para as operações.
8. `salacheia(cinema, filme)` - diz se a sala em que o filme está a ser exibido tem lotação esgotada.
9. `filmedispo(cinema, filme)` - dá como resultado **True** se o filme está em exibição no cinema e **False** caso contrário.
10. `lugocupados(cinema, movie)`- dá como resultado uma lista dos lugares ocupados (bilhetes vendidos) para o filme a ser exibido

````python
# TPC4
# Modelo dos cinemas
# Cinema = [Sala]
# Sala = (nlugares, Vendidos, filme)
# nlugares = Int
# filme = String 
# Vendidos = [Int]
cinema1 = [(150, [25,17,30], "Top Gun: Maverick"),(2, [1, 2], "Mínimos 2"), (100, [5], "Avatar")]

def listar(cinema):
    print("Em cartaz:")
    for sala in cinema:
        nlugares, Vendidos, filme = sala
        print(filme)
    return


def disponivel(cinema, film, lugar):
    res = False
    for sala in cinema:
        nlugares, Vendidos, filme = sala
        if film == filme and lugar <= nlugares:
            if lugar not in Vendidos:
                res = True
    return res



def vendebilhete(cinema, film, lugar):
    if disponivel(cinema, film, lugar):
        for sala in cinema:
            nlugares, Vendidos, filme = sala
            if film ==filme:
                Vendidos.append(lugar)
    return cinema



def listardisponibilidades(cinema):
    i = 1
    for sala in cinema:
        nlugares, Vendidos, filme = sala
        lugdispo = nlugares - len(Vendidos)
        print("Sala", i, ":", filme, "|", lugdispo, "lugares disponíveis")
        i = i + 1

def criarsalanova():
    lugares = int(input("Nr de lugares:"))
    ocupados = []
    movie = input("Filme a exibir:")
    salanova = lugares, ocupados, movie
    return salanova

def inserirsala(cinema, salanova):
    existe = False
    for sala in cinema:
        nlugares, Vendidos, filme = sala
        if salanova[-1] == filme:
            existe = True
    if existe == False:
        cinema.append(salanova)
    return cinema


def salacheia(cinema, movie):
    continua = True
    for sala in cinema:
        nlugares, Vendidos, filme = sala
        if continua == True:
            if filme == movie:
                if nlugares == len(Vendidos):
                    res = print("Lotação esgotada")
                elif nlugares > len(Vendidos):
                    res = print("Ainda tem lugares")
                continua = False
    if filme!= movie and continua == True:
        res = print("'Erro'")
    return res


def filmedispo(cinema, movie):
    dispo = False
    for sala in cinema:
        nlugares, Vendidos, filme = sala
        if filme == movie:
            dispo = True
    return dispo

def lugocupados(cinema, movie):
    v = False 
    for sala in cinema:
       nlugares, Vendidos, filme = sala
       if filme == movie:
            v = True
            ocupados = Vendidos
    if v == False:
        ocupados = print("'Erro'")   
    return ocupados


def appcinema():
    menu ="""...........................................................................................................
1) Filmes em cartaz | 2) Ver disponibilidade de lugares | 3) Comprar bilhete | 4) Disponibilidade das salas 
5) Acrescentar sala | 6) Lotação das salas | 7) Pesquisar filmes | 8) Bilhetes vendidos | 0) Terminar 
..........................................................................................................."""
    print(menu) 
    option = int(input("Selecione uma opção"))
    while option in range(1,9):
        if option == 1:
            listar(cinema1)
        if option == 2:
            film = input("Que filme pretende ver?")
            lugar = int(input("Que lugar quer?"))
            print(disponivel(cinema1, film, lugar))
        if option == 3:
            film = input("Que filme pretende ver?")
            lugar = int(input("Que lugar quer?"))
            print(vendebilhete(cinema1, film, lugar))
        if option == 4:
            listardisponibilidades(cinema1)
        if option == 5:
            v = criarsalanova()
            print(inserirsala(cinema1, v))
        if option == 6:
            movie = input("Lotação da sala com o filme...")
            salacheia(cinema1, movie)
        if option == 7:
            movie = input("O filme... está em cartaz?")
            filmedispo(cinema1, movie)
        if option == 8:
            movie = input("Bilhetes vendidos para o filme...")
            lugocupados(cinema1, movie)
        print(menu)
        option = int(input("Selecione uma opção"))
    if option == 0:
        print("Terminou")

appcinema()
````
